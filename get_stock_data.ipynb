#### **5. TỶ GIÁ NGOẠI TỆ USD VÀ VNĐ**

```python
fx = Vnstock().fx(symbol='USDVND', source='MSN')
start = '2014-11-16'
end = '2024-06-30'
df = pd.DataFrame()

while pd.to_datetime(end) > pd.to_datetime(start):
    temp = fx.quote.history(start=start, end=end).reset_index(drop=True)
    end = (pd.to_datetime(temp.time[0]) - pd.Timedelta(days=1)).strftime('%Y-%m-%d')
    temp = temp.sort_values(by='time', ascending=False).reset_index(drop=True)
    df = pd.concat([df, temp]).reset_index(drop=True)

df
```
